**Question 1: What is the difference between descriptive statistics and inferential statistics? Explain with examples.**


>**1. Descriptive Statistics**

>>Descriptive statistics is concerned with summarizing, organizing, and presenting data in a way that makes it easier to understand. It doesn't make conclusions about a larger population but focuses on the data at hand.

>**Key Features:**

>>**Central Tendency:** Measures the center of the data (mean, median, mode).


>>**Dispersion:** Measures the spread of the data (range, variance, standard deviation).


>>**Data Visualization:** Tools like histograms, pie charts, and box plots to summarize data visually.


>**Examples:**

>>**Average Test Scores:** If you have the test scores of 30 students in a class, you might compute the mean (average), median, or mode of the scores to describe the overall performance of the class.


>>**Age Distribution:** You could calculate the average age, standard deviation, and create a histogram to show the age distribution of a group of people.


>>>**For example:**

>>>If you have a dataset of 10 exam scores: [85, 90, 88, 76, 92, 78, 85, 93, 80, 87]

>>>The mean is 86.4 (sum of all scores divided by the number of scores).


>>>The range is 17 (the difference between the highest and lowest scores: 93 - 76).


>>>You could also create a histogram to visually display how scores are distributed.


>**2. Inferential Statistics**

>>Inferential statistics goes beyond summarizing data; it uses data from a sample to make generalizations or predictions about a larger population. It relies on probability theory to draw conclusions that are not directly observable.

>**Key Features:**

>>**Hypothesis Testing:** Involves testing an assumption (e.g., t-tests, chi-square tests).


>>**Confidence Intervals:** Estimates a range of values within which the true population parameter is likely to lie.


>>**Regression and Correlation:** Used to model relationships between variables and make predictions.


>**Examples:**

>>**Polls and Surveys:** In political polls, you might survey 1,000 people to infer the voting preferences of a population of 100 million. Here, inferential statistics helps generalize the preferences of the entire population based on the sample.


>>**Product Testing:** A company might test a new product on a small sample of customers and use inferential statistics to predict how the product will perform with the entire market.


>>>**For example:**

 >>>Let’s say a survey is done on 200 people out of a population of 1,000,000 to determine the proportion of people who like a particular brand. If 75% of the sample likes the brand, inferential statistics can be used to estimate that the population proportion is about 75%, with a certain level of confidence (e.g., 95%).

>**Key Differences**

>>**Purpose:**


>>Descriptive statistics summarizes and describes the data by providing measures like the mean, median, or range. It gives a clear picture of the dataset at hand.


>>Inferential statistics makes predictions or inferences about a larger population based on sample data. It helps generalize beyond the data available.


>>**Data Type:**


>>Descriptive statistics focuses on the dataset at hand, providing insights based solely on the collected data.


>>Inferential statistics uses sample data to draw conclusions about a broader population, allowing for estimations and generalizations.


>>>**Example:**


>>>Descriptive statistics might involve calculating the mean, median, mode, or standard deviation to summarize a dataset, such as finding the average score of a class.


>>>Inferential statistics would be used to estimate a population’s mean, perform hypothesis testing, or calculate confidence intervals to make predictions about the larger group.


>>**Generalization:**


>>In descriptive statistics, there is no generalization beyond the sample. It only describes the data that has been collected.


>>In inferential statistics, generalizations are made from the sample to the larger population, allowing for broader conclusions or decisions.


>>**Tools Used:**


>>Descriptive statistics relies on tools like the mean, median, mode, range, variance, and various graphs (such as histograms or pie charts) to present and summarize data.


>>Inferential statistics involves more advanced techniques like T-tests, chi-square tests, regression analysis, and the use of confidence intervals to draw conclusions about the population.








**Question 2: What is sampling in statistics? Explain the differences between random and stratified sampling.**

>**Sampling in Statistics:**

>Sampling in statistics refers to the process of selecting a subset (or sample) from a larger population to make inferences or generalizations about the entire population. The idea is to draw conclusions based on a representative sample rather than gathering data from every individual in the population, which is often impractical or too costly.
There are different types of sampling methods, but the goal is always to obtain a sample that accurately reflects the characteristics of the population.

>Two Common Types of Sampling: **Random and Stratified Sampling**

>**1. Random Sampling:**

>>**Definition:**

 >>Random sampling is a technique where every individual in the population has an equal chance of being selected for the sample. It is considered one of the most basic and unbiased sampling methods.

>>**How it works:**

>>In random sampling, you randomly select members from the population, either using random number generators, drawing names from a hat, or any other method that ensures each member of the population has an equal chance of being chosen.


>>**Example:**

>>Imagine you have a population of 500 students in a school, and you need to select 50 students to survey about their favorite sports. If you randomly choose 50 students, every student has the same probability of being selected, regardless of any specific characteristics (like age, grade, or gender).

>>**Advantages:**

>>Simple and easy to implement.


>>Reduces bias because each member has an equal chance of being selected.


>>**Disadvantages:**

>>Can still result in a sample that is not perfectly representative of the population if the population is heterogeneous (diverse). For example, if there are subgroups within the population that are not evenly distributed, random sampling might end up over- or under-representing certain groups.


>**2. Stratified Sampling:**

>>**Definition:**

 >>Stratified sampling involves dividing the population into different subgroups or "strata" that share a common characteristic (such as age, gender, income, or education level). Then, random sampling is performed within each subgroup. This ensures that each subgroup is well-represented in the final sample.

>>**How it works:**

>>First, you divide the population into distinct strata based on a characteristic of interest (e.g., age groups, income levels, or geographical locations).


>>Then, you randomly select members from each of the strata. The number of samples chosen from each stratum can be proportional to the size of that stratum in the population or chosen equally across strata.


>>**Example:**

 >>Let’s say you are studying the opinions of students in a university with 3 different departments: Science, Arts, and Business. If you randomly sample from the entire student body without considering the number of students in each department, you might end up with too many students from one department and too few from another. With stratified sampling, you divide the student body into 3 strata based on their department, then randomly sample from each department, ensuring all departments are proportionally represented.

>>**Advantages:**

>>Ensures that specific subgroups of the population are represented in the sample.


>>Can lead to more precise estimates, particularly if the variable you're studying varies across strata.


>>**Disadvantages:**

>>More complex and time-consuming than random sampling.


>>Requires knowledge of the population structure in advance (i.e., the different strata must be known).

>**Key Differences Between Random and Stratified Sampling:**

>>**Definition:**


>>In Random Sampling, every individual in the population has an equal chance of being selected, which means the selection is purely based on chance.


>>Stratified Sampling involves dividing the population into subgroups, or strata, based on a shared characteristic, and then performing random sampling within each stratum to ensure diversity within the sample.


>>**Sampling Method:**


>>With Random Sampling, you randomly select individuals from the entire population without considering any subgroup characteristics.


>>Stratified Sampling requires you to first divide the population into distinct strata and then randomly sample individuals from each subgroup, ensuring each group is represented.


>>**Bias:**


>>Random Sampling may result in bias if the population is heterogeneous because some subgroups may be underrepresented or overrepresented in the sample due to chance.


>>Stratified Sampling reduces this potential bias by ensuring that every subgroup of the population is represented proportionally in the sample.


>>**Population Knowledge:**


>>Random Sampling does not require any prior knowledge of specific subgroups within the population, making it simple and straightforward to implement.


>>Stratified Sampling, on the other hand, requires detailed knowledge of the population structure, as you need to identify and define the relevant strata before sampling.


>>**Representativeness**:


>>In Random Sampling, there's a risk that some subgroups might be underrepresented or overrepresented, leading to an unbalanced representation of the population.


>>Stratified Sampling guarantees that all important subgroups are proportionally represented in the final sample, improving the accuracy of the inferences.


>>**Usefulness:**


>>Random Sampling is ideal when the population is relatively homogeneous, meaning the differences between individuals are small, or when no particular subgroup needs special representation.


>>Stratified Sampling is most useful when the population is heterogeneous and contains distinct subgroups that must be represented accurately in the sample, such as different age groups or income levels.


>>**Complexity:**


>>Random Sampling is relatively simple to implement because it involves no additional steps beyond random selection.


>>Stratified Sampling is more complex and time-consuming because it requires dividing the population into meaningful strata, followed by random sampling within each of those strata.





**Question 3: Define mean, median, and mode. Explain why these measures of central tendency are important.**

>**Measures of Central Tendency:**

>The mean, median, and mode are three key measures of central tendency in statistics. They help summarize a dataset by identifying a single value that represents the center or typical value of a distribution.

>**1. Mean (Arithmetic Average):**

>>**Definition:**

 >>The mean is the sum of all the values in a dataset, divided by the number of values. It represents the "average" value.

>>>**Formula:   Mean=∑(all values in the dataset) ​/ number of values**

>>**Example:**

>> For the dataset [5, 8, 12, 20, 25], the mean would be calculated as:
>>Mean = [5+8+12+20+25​] / 5​ = 70/5 = 14

>>So, the mean is **14**.

>**2. Median:**

>>**Definition:**

>> The median is the middle value in a dataset when the values are arranged in ascending (or descending) order. If the dataset has an odd number of values, the median is the middle value. If there is an even number of values, the median is the average of the two middle values.

>>**Example:**

>> For the dataset [5, 8, 12, 20, 25], the median is 12, because it’s the middle value when the data is ordered.

>>For the dataset [5, 8, 12, 20], the median would be:

>> Median = [8+12]​ / 2 =10

>>So, the median is **10**.

>**3. Mode:**

>>**Definition:**

>> The mode is the value that appears most frequently in the dataset. A dataset can have no mode (if all values are unique), one mode (unimodal), or multiple modes (bimodal or multimodal) if two or more values appear with the same highest frequency.

>>**Example:**

>> For the dataset [5, 8, 12, 12, 25], the mode is 12, as it appears twice, while all other values appear only once.
For the dataset [5, 8, 12, 12, 25, 25], this dataset is bimodal because both 12 and 25 appear twice.

>**Importance of Measures of Central Tendency:**

>>These measures are important because they provide a quick and concise summary of a dataset, helping us understand the general pattern or center of the data. Here’s why each one is useful:

>>**1. Mean:**

>>**Importance:** The mean is widely used because it incorporates all values in the dataset, providing a comprehensive measure of central tendency.


>>**When to Use:** It is most useful when the data is symmetrically distributed and when there are no extreme outliers that could skew the results.


>>**Example:** Calculating the average salary in a company or average exam score in a class.


>>**2. Median:**

>>**Importance:** The median is important because it is resistant to outliers and skewed data. In cases where extreme values can distort the mean, the median gives a better representation of the central value.


>>**When to Use:** It’s especially useful when the data is skewed or contains outliers. For instance, in income data, where a few high earners could distort the mean, the median provides a more accurate reflection of the "typical" income.


>>**Example:** If you want to find the median house price in an area where most homes are affordable, but a few mansions skew the average.


>>**3. Mode:**

>>**Importance:** The mode is helpful for identifying the most frequent or common value in a dataset, which can be useful in fields like marketing, retail, or psychology, where knowing the most popular item, response, or behavior is important.


>>**When to Use:** It is useful for categorical data (like most common color, size, or brand preference), and when identifying trends or patterns is important.


>>**Example:** In a survey, if you want to know the most preferred brand of shoes, the mode would give you the most frequently chosen brand.







**Question 4: Explain skewness and kurtosis. What does a positive skew imply about the data?**

>**Skewness and Kurtosis:**

>Both skewness and kurtosis are statistical measures that describe the shape of a distribution. While skewness measures the asymmetry, kurtosis measures the "tailedness" or the peakedness of the distribution.

>**1. Skewness:**

>**Definition:**

 >Skewness refers to the degree of asymmetry or departure from symmetry in a dataset or distribution. It indicates whether the data tends to be more concentrated on one side of the mean.

>**Types of Skewness:**

>**Positive Skew (Right Skew):**

 >If the distribution has a long right tail, it is said to be positively skewed. The mean is usually greater than the median in such distributions because the tail pulls the mean to the right.

>>**Example:** Income distribution often shows positive skew, where most people earn average or low incomes, but there are a few individuals with extremely high incomes that pull the mean to the right.


>**Negative Skew (Left Skew):**

 >If the distribution has a long left tail, it is said to be negatively skewed. In this case, the mean will usually be less than the median, because the tail pulls the mean to the left.

>>**Example:** The age of retirement in a population might exhibit negative skew, where most people retire around a certain age, but a few retire much earlier, pulling the mean to the left.


>**Zero Skew (Symmetrical Distribution):**

> A skewness value of 0 indicates a perfectly symmetrical distribution, where the left and right sides are mirror images of each other (e.g., a normal distribution).


>**What does a Positive Skew imply?**

>>**A positive skew (right skew) means:**

>>Most of the data points are clustered on the left side of the distribution, with the tail extending to the right.


>>The mean is typically greater than the median because the tail on the right side pulls the average higher.


>>It often indicates the presence of outliers on the right-hand side (higher values).


>>>**Example of Positive Skew:**

 >>>Consider a dataset representing the number of years of experience in a job for a group of employees:

>>>Most employees may have 1–5 years of experience, but there are a few employees with 20+ years of experience. The mean would be pulled toward the higher end, resulting in a positive skew.



>**2. Kurtosis:**

>**Definition:**

 >Kurtosis refers to the shape of the distribution's tails and its peak. It measures the "tailedness" of a distribution, indicating how much of the data is in the extremes (the tails) versus near the center.

>**Types of Kurtosis:**

>**Leptokurtic:**

> If a distribution has heavy tails and a sharp peak, it is said to have positive kurtosis. This distribution has more extreme values (outliers) than a normal distribution.


>>**Example:** A stock market return distribution might be leptokurtic, where there are a few extreme gains or losses compared to the typical returns.


>**Platykurtic:**

 >If the distribution is flat with light tails, it is said to have negative kurtosis. This distribution has fewer outliers and more evenly distributed values.


>>**Example:** A dataset on the height of a group of people might be platykurtic because there are fewer extreme heights, and the data is more spread out around the average.


>**Mesokurtic:**

> A normal distribution has zero kurtosis, meaning it has a moderate peak and typical tails. The tails of a normal distribution are neither too heavy nor too light.


>>**Example:** A dataset that follows a normal distribution, such as test scores for a well-constructed exam.


**Question 5: Implement a Python program to compute the mean, median, and mode of a given list of numbers.**

**numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]**


```python
import statistics


# List of numbers
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]


# Compute the mean, median, and mode
mean = statistics.mean(numbers)
median = statistics.median(numbers)
mode = statistics.mode(numbers)


# Display the results
print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)


```

>**Explanation:**

>>**Mean:** The average value of the numbers. It's calculated as the sum of all numbers divided by the total count. 19.47 (rounded to 2 decimal places).

>>**Median:** The middle value of the sorted numbers. If the number of elements is odd, it's the middle number; if even, it's the average of the two middle values. 19 (the middle value when sorted).

>>**Mode:** The number that appears most frequently in the list. 12 (the most frequent value).


**Question 6: Compute the covariance and correlation coefficient between the** **following two datasets provided as lists in Python:**

**list_x = [10, 20, 30, 40, 50]**

**list_y = [15, 25, 35, 45, 60]**


```python
import numpy as np


# Provided datasets
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]


# Computing covariance
cov_matrix = np.cov(list_x, list_y)
cov_xy = cov_matrix[0, 1]


# Computing correlation coefficient
corr_coef = np.corrcoef(list_x, list_y)[0, 1]


print(f"Covariance: {cov_xy}")
print(f"Correlation coefficient: {corr_coef}")
```

**Question 7: Write a Python script to draw a boxplot for the following numeric list and identify its outliers. Explain the result:**

**data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]**


```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


# Data list
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]


# Create a boxplot
sns.boxplot(data=data)


# Display the plot
plt.title("Boxplot of the Given Data")
plt.show()


# Calculate the quartiles and outliers
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1


lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR


# Identify outliers
outliers = [x for x in data if x < lower_bound or x > upper_bound]


# Output the result
print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")
print(f"Outliers: {outliers}")
```


**Question 8: You are working as a data analyst in an e-commerce company. The marketing team wants to know if there is a relationship between advertising spend and daily sales.**

**● Explain how you would use covariance and correlation to explore this relationship.**

**● Write Python code to compute the correlation between the two lists:**

**advertising_spend = [200, 250, 300, 400, 500]**

**daily_sales = [2200, 2450, 2750, 3200, 4000]**

**Answer:**

>**Covariance**

>Covariance is a numerical value that indicates how two variables change together. In your case:

>>If advertising spend increases, and daily sales also increase, then covariance will be positive.


>>If one increases and the other decreases, covariance will be negative.


>>If they change independently, covariance will be close to zero.


>>**Limitation:**

 >>Covariance gives the direction (positive or negative) of a relationship, but not the strength, because the magnitude of covariance depends on the units of the data.

>**Correlation**

>Correlation is a standardized form of covariance, making it much easier to interpret.
>It ranges between -1 and 1.


>A value of 1 means a perfect positive linear relationship.


>A value of -1 means a perfect negative linear relationship.


>A value close to 0 means no linear relationship.


>**Why use correlation?**

>It normalizes the covariance.


>It tells you how strong the relationship is, and whether it's positive or negative.


>It's unitless, so you can compare it across different datasets.



```python
import numpy as np


# Advertising and sales data
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]


# 1. Calculate Covariance Matrix
cov_matrix = np.cov(advertising_spend, daily_sales)
covariance = cov_matrix[0, 1]


# 2. Calculate Correlation Coefficient
correlation_matrix = np.corrcoef(advertising_spend, daily_sales)
correlation = correlation_matrix[0, 1]


# Output the results
print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)
```

>**Interpretation**

>>**Covariance = 84,875.0**

>>This is a positive number, which confirms that:


 >>As advertising spend increases, daily sales also increase.



>>The magnitude (84,875) itself isn’t directly interpretable unless you compare it with other datasets, because it depends on the scale/units of your variables (dollars, units sold, etc.).



>>**Correlation Coefficient ≈ 0.994**

>>This is very close to 1, meaning:


>>There is a very strong positive linear relationship.


>**In real-world terms:** When your company spends more on advertising, you can almost certainly expect higher daily sales.


>Since it's over 0.99, this is an almost perfect linear correlation in statistical terms.



>**Final Statement**


>**I analyzed the relationship between advertising spend and daily sales using statistical methods. The results show a very strong positive correlation (r ≈ 0.994). This means that increased advertising spend is closely associated with higher daily sales. It’s a strong indication that our advertising investments are effective, and further investment could potentially lead to increased revenue.**


**Question 9: Your team has collected customer satisfaction survey data on a scale of 1-10 and wants to understand its distribution before launching a new product**.

>**● Explain which summary statistics and visualizations (e.g. mean, standard deviation, histogram) you’d use.**

>**● Write Python code to create a histogram using Matplotlib for the survey data:**

>**survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]**


**Answer:**

>**Summary Statistics**

>These give you a quick overview of the data:

>>**Mean:** Average satisfaction score (center of the data)


>>**Median:** Middle value (less affected by outliers)


>>**Mode:** Most common score (shows what most people selected)


>>**Standard Deviation:** How spread out the scores are


>>**Min/Max:** To understand the range


>These help you answer questions like:

>Are most customers rating satisfaction high? Is there a lot of variability?

>**Visualizations**

>Visual tools help identify patterns, skewness, or outliers.

>>**Histogram:** Best for seeing the distribution (e.g., is it bell-shaped, skewed, flat?)


>>**Boxplot (optional):** Shows median, quartiles, and potential outliers


>>**Bar chart (for categorical view):** If you want to treat scores as distinct categories (1 to 10)




```python
import matplotlib.pyplot as plt
import numpy as np


# Survey scores
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]


# Summary statistics
mean = np.mean(survey_scores)
median = np.median(survey_scores)
std_dev = np.std(survey_scores)
mode = max(set(survey_scores), key=survey_scores.count)


print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
print(f"Standard Deviation: {std_dev:.2f}")
print(f"Min: {min(survey_scores)}, Max: {max(survey_scores)}")


# Create histogram
plt.hist(survey_scores, bins=range(1, 12), edgecolor='black', align='left', color='skyblue')
plt.title("Customer Satisfaction Score Distribution")
plt.xlabel("Satisfaction Score (1-10)")
plt.ylabel("Frequency")
plt.xticks(range(1, 11))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

>**Explanation:**

>>**Scores are centered around 7**, with several 8s and 9s →** skewed slightly right (positive bias).**


>>The **low standard deviation (1.58)** tells you that** most responses are clustered, not wildly spread.**


>>The **minimum score of 4** is low but not extreme — **possibly a few outliers but not too concerning.**


>**The average customer satisfaction score is 7.33, with a median and mode of 7. The distribution is tightly clustered, with a standard deviation of just 1.58, indicating consistent customer experiences. Most scores fall between 6 and 9, suggesting a strong overall sentiment with room for small improvements. This favorable distribution supports a confident new product launch, especially if we address feedback from the few lower-scoring customers.**
